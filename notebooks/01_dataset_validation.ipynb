## 1. Contexto do dataset

O dataset utilizado vem do **NASA Exoplanet Archive**, uma base mantida pela NASA/IPAC que reúne dados relacionados a exoplanetas, candidatos planetários e observações astronômicas.

Neste projeto, será utilizada a tabela **Kepler Objects of Interest (KOI)**. Um KOI é um objeto observado pela missão Kepler que apresentou sinais compatíveis com a possível presença de um exoplaneta, geralmente por meio do método de trânsito planetário.

A tabela contém objetos classificados em categorias como:

- `CONFIRMED`: planeta confirmado;
- `CANDIDATE`: candidato a exoplaneta;
- `FALSE POSITIVE`: sinal que parecia indicar um planeta, mas foi classificado como falso positivo.

Essa estrutura é adequada para um problema de classificação supervisionada, pois permite treinar modelos para distinguir candidatos promissores de falsos positivos com base em características observacionais.

### Variáveis físicas e observacionais encontradas

Nesta parte do dataset aparecem variáveis mais diretamente ligadas ao problema de identificação de exoplanetas.

Entre elas, temos `koi_prad`, que representa o raio estimado do possível planeta, `koi_duration`, que indica a duração do trânsito, `koi_depth`, que representa a profundidade da queda de brilho, e `koi_period`, que indica o período orbital estimado.

Também aparecem variáveis relacionadas à estrela hospedeira, como `koi_steff`, `koi_slogg`, `koi_srad` e `koi_smass`. Essas informações são relevantes porque o sinal observado depende não apenas do possível planeta, mas também das características da estrela ao redor da qual ele orbita.

Essas colunas reforçam que o dataset tem atributos suficientes para construir modelos de classificação e, posteriormente, investigar quais características ajudam a diferenciar candidatos, planetas confirmados e falsos positivos.

### Colunas com bom preenchimento

A análise das colunas com menos valores ausentes mostra que algumas variáveis importantes estão completas ou quase completas no dataset.

Entre elas, estão `koi_disposition`, que será usada como variável-alvo, além de variáveis como `koi_period` e `koi_duration`, que descrevem características do trânsito observado.

Também existem colunas de identificação e posição, como `kepid`, `kepoi_name`, `ra` e `dec`, que ajudam a localizar e reconhecer os objetos, mas não necessariamente serão usadas como variáveis principais para o modelo.

Um ponto de atenção é a presença de colunas como `koi_fpflag_ec`, `koi_fpflag_co`, `koi_fpflag_ss` e `koi_fpflag_nt`. Essas colunas parecem estar diretamente relacionadas à identificação de falsos positivos. Por isso, elas devem ser analisadas com cuidado para evitar **data leakage**, ou seja, vazamento de informação da resposta para as variáveis de entrada. Se isso acontecer, o modelo pode usar atalhos e apresentar uma avaliação artificialmente boa.

Com isso, a próxima etapa será selecionar um primeiro conjunto de variáveis candidatas para análise, priorizando colunas com bom preenchimento, significado relevante para o problema e baixo risco de vazamento de dados.

```python
selected_columns = [
    # Identificação
    "kepid",
    "kepoi_name",

    # Variável-alvo
    "koi_disposition",

    # Características do trânsito
    "koi_period",
    "koi_duration",
    "koi_depth",
    "koi_num_transits",
    "koi_model_snr",

    # Características estimadas do possível planeta
    "koi_prad",
    "koi_teq",
    "koi_insol",

    # Características da estrela hospedeira
    "koi_steff",
    "koi_slogg",
    "koi_srad",
    "koi_smass",

    # Observação/brilho
    "koi_kepmag",

    # Posição no céu
    "ra",
    "dec"
]

df_selected = df[selected_columns].copy()

df_selected.head()
```

### Análise da seleção inicial de variáveis

O novo dataframe reúne as colunas mais importantes para a primeira versão da análise. Foram mantidas variáveis de identificação, a variável-alvo `koi_disposition` e características físicas e observacionais relacionadas ao trânsito, ao possível planeta e à estrela hospedeira.

Essa seleção ajuda a reduzir o dataset para um conjunto mais interpretável, sem perder informações relevantes para o problema. Nas próximas etapas, ainda vamos verificar valores nulos, tipos de dados e possíveis ajustes antes de definir as variáveis finais para modelagem.

```python
selected_missing = (
    df_selected.isna()
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

selected_missing.columns = ["column_name", "missing_count"]

selected_missing["missing_percentage"] = (
    selected_missing["missing_count"] / len(df_selected) * 100
).round(2)

selected_missing
```

### Análise dos valores ausentes nas variáveis selecionadas

A verificação dos valores ausentes mostra que as variáveis selecionadas possuem um nível de preenchimento adequado para a continuidade do projeto.

A coluna com maior quantidade de valores ausentes é `koi_num_transits`, com 11,94% dos registros vazios. Apesar disso, a variável ainda pode ser útil, pois representa a quantidade de trânsitos observados e pode contribuir para avaliar a consistência do sinal detectado.

Outras variáveis importantes, como `koi_depth`, `koi_model_snr`, `koi_prad`, `koi_steff`, `koi_slogg`, `koi_srad` e `koi_smass`, possuem cerca de 3,80% de valores ausentes. Esse percentual é baixo o suficiente para ser tratado na etapa de limpeza, sem a necessidade de remover essas colunas neste momento.

Com isso, a seleção inicial de variáveis continua válida. Nas próximas etapas, será necessário decidir a melhor estratégia de tratamento para esses valores, como remoção de linhas incompletas ou preenchimento por alguma medida estatística, processo conhecido como `imputation`.
